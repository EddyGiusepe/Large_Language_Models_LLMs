# <font color="red">Contextualizando</font>

No domínio do Processamento de Linguagem Natural (`NLP`), resumir documentos extensos ou múltiplos apresenta um desafio formidável. O grande volume de dados geralmente leva a tempos de processamento mais lentos e restrições de memória, exigindo investimentos em infraestrutura de computação de alto desempenho. No entanto, o advento do `Langchain` oferece uma solução inovadora ao dividir documentos grandes em partes menores. Ao empregar processamento simultâneo ou serial, com base no tipo de `cadeia` (chain) escolhido, `o Langchain elimina efetivamente o fardo de lidar com a emissão máxima de tokens`.

Além dos detalhes de lidar com grandes quantidades de informações, combinar insights de vários documentos para criar um resumo coeso representa outro obstáculo. Discrepâncias de terminologia, dados conflitantes e aspectos variados do tópico dificultam a fusão efetiva do conteúdo. Felizmente, `Langchain` vem em socorro. Ao armazenar informações relevantes de documentos anteriores no atual, ele estabelece uma cadeia abrangente de documentos interconectados. Essa abordagem inteligente permite que os modelos de resumo de NLP contextualizem as informações, comprovem sua importância e mantenham a ordem adequada das sentenças no conteúdo resumido.

# <font color="red">O que é LangChain</font>

`LangChain`, uma estrutura inovadora desenvolvida por `Harrison Chase`, surgiu como um divisor de águas no domínio dos aplicativos de modelo de linguagem grande (`LLM`). Desde seu lançamento inicial como um projeto de `código aberto` em outubro de `2022`, o LangChain atraiu atenção significativa, acumulando impressionantes `41900 estrelas no GitHub` e atraindo uma comunidade vibrante de mais de 800 colaboradores.

`LangChain` atua como uma ponte entre modelos `LLM`, como `OpenAI` e `HuggingFace Hub`, e fontes de dados externas como `Google`, `Wikipedia`, `Notion` e `Wolfram`. Ao conectar perfeitamente esses recursos, o LangChain capacita os desenvolvedores a aproveitar uma variedade de abstrações e ferramentas. Isso inclui `cadeias` e `agentes`, que fornecem fluxos de trabalho estruturados, bem como modelos de `prompt`, gerenciamento de memória, carregadores de documentos e analisadores de saída, todos os quais facilitam a interação suave entre entrada e saída de texto.



# <font color="red">Componentes do LangChain</font>

Os sete módulos principais do `LangChain` estão listados abaixo:

* `Modelos:` Integração de LLMs de código aberto ou fechado

* `Prompts:` Entrada de usuário baseada em modelo e formatação de saída para modelos LLM

* `Índices:` Estruturação e preparação de dados para interação ideal com modelos LLM

* `Memória:` permitindo que cadeias ou agentes retenham interações de curto e longo prazo com os usuários

* `Cadeias:` Combinando vários componentes ou cadeias em um único pipeline para processamento simplificado

* `Agentes:` Entidades de tomada de decisão que utilizam ferramentas e dados disponíveis com base na entrada

* `Callbacks:` funções acionadas para executar ações específicas durante a execução do LLM

# <font color="red">Cadeias de sumarização</font>

LangChain fornece várias cadeias para resumir o conteúdo. As cadeias mais usadas são `map_reduce`, `stuff` e `refine`.

`map_reduce`

A cadeia `map_reduce` aproveita o poder dos `prompts` iniciais em blocos (`chunks`) de dados individuais para gerar resumos ou respostas com base apenas em seções específicas de um documento. Ele vai um passo além ao empregar um prompt distinto que combina as saídas iniciais, resultando em um resumo ou resposta abrangente e coeso que abrange todo o documento. Essa abordagem mostra sua capacidade de lidar sem esforço até mesmo com os documentos mais extensos e intrincados, fornecendo resultados eficientes e precisos.

`stuff`

O método `stuffing` envolve a inclusão de todos os dados relevantes como contexto no `prompt`, permitindo que sejam passados ​​para o modelo de linguagem. É uma abordagem direta que funciona de forma eficaz para lidar com uma parte menor dos dados. No entanto, à medida que o número de partes de dados aumenta, esse método se torna impraticável e inadequado.

`refine`

O método de refinamento (`refine`) começa com um prompt inicial no primeiro bloco (`chunk`) de dados para gerar uma saída. Essa saída é então passada junto com o documento subsequente para o modelo de linguagem, instruindo-o a refinar a saída incorporando as novas informações do documento. Esse processo iterativo permite a melhoria gradual e o aprimoramento da saída à medida que mais documentos são processados.
