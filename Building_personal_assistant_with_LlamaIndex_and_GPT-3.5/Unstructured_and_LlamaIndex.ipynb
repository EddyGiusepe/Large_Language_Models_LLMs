# Contextualizando

Neste tutorial, mostraremos como obter `insights` facilmente dos arquivos `SEC 10-K`, usando o poder de alguns componentes principais: 

<font color="red">1)</font> Large Language Models (`LLM's`), 

<font color="red">2)</font> Análise de dados por meio de `Unstructured` e 

<font color="red">3)</font> Indexação de Dados através do `LlamaIndex!` 


Mostramos como a ferramenta LlamaIndex pode ajudá-lo a construir uma interface de consulta incrível sobre seus próprios dados.

Os `Large Language Models (LLM's)` estão começando a revolucionar a forma como os usuários podem pesquisar, interagir e gerar novos conteúdos. A incrível popularidade de ferramentas como o `ChatGPT` serviu apenas para acelerar a descoberta de valor em todas essas dimensões. Simplesmente inserindo uma query de pesquisa em uma caixa de texto, os usuários podem interagir com informações de uma incrível variedade de fontes de dados, incorporadas à rede. Além disso, o valor dos `LLMs` não é simplesmente que os usuários possam acessar essas informações para fins de pesquisa e recuperação; os usuários têm a capacidade de aproveitar essas informações para uma variedade de `tarefas` de geração e raciocínio, desde respostas a perguntas destiladas, resumos, redação de conteúdo de formato longo, transferência de estilo e até mesmo tomada de decisões + execução de ações.
